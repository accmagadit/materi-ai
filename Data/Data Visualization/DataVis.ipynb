```python
import numpy as np #fungsi matematika
import pandas as pd #load dataset csv, json dll
```

```python
url = 'https://people.sc.fsu.edu/~jburkardt/data/csv/freshman_kgs.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.describe()
```

```python
df.describe(include='object')
```

<h2>Visualisasi pakai matplollib</h2>

```python
import matplotlib.pyplot as plt
```

```python
#menghitung jumlah pria dan wanita
jumlahPria = df['Sex'].value_counts()['M']
jumlahWanita = df['Sex'].value_counts()['F']

#membuat bar graph
plt.bar(['Pria','Wanita'], [jumlahPria,jumlahWanita])
```

kita dapat mengubah ukuran dan menambah keterangan lain dalam plot, seperti
- judul plot
- nama sumbu x
- nama sumbu y
- warna barchart

```python
plt.figure(figsize=(10,3))
plt.bar(['Pria','Wanita'],[jumlahPria,jumlahWanita],color=['orange','pink'])
plt.title('Perbandingan jumlah pria dan wanita')
plt.xlabel('kategori')
plt.ylabel('jumlah')
```

```python
#menghitung rata-rata berat pria dan wanita bulan september
rataBeratPria = df[df['Sex'] == 'M'][' "Weight (Sep)"'].mean()
rataBeratWanita = df[df['Sex'] == 'F'][' "Weight (Sep)"'].mean()
weights = [rataBeratPria,rataBeratWanita]
labels = ['Pria','Wanita']

plt.figure(figsize=(7,8))
plt.bar(labels,weights)
plt.title('erbandingan rata rata berat badan pria dan wanita bulan september')
plt.show()
```

<h2>Histogram</h2>
biasa digunakan untuk menampilkan distribusi frekuensi.

```python
plt.hist(df[' "Weight (Sep)"'])
plt.title('Histogram dari berart badan bulan april')
plt.ylabel('Frekuensi')
plt.xlabel('Berat')
plt.show()
```

Cara lain membuat histogram

```python
df[' "Weight (Apr)"'].plot(kind='hist', figsize=(7,5))
plt.title('Fistogram dari berat badan bulan april')
plt.ylabel('Frekuensi')
plt.xlabel('berat')
plt.show()
```

<h2>Line Chart</h2>

```python
bmi_april = df[' "BMI (Sep)"']

plt.plot(bmi_april)
plt.title('Perbandingan BMI bulan April Dan September')
plt.xlabel('Peserta')
plt.ylabel('BMI')
plt.show()
```

Contoh line chart untuk membandingkan data

```python
weights_september = df[' "Weight (Sep)"']
bmi_september = df[' "BMI (Sep)"']

plt.plot(weights_september,'o--' ,label='Berat Bulan September')
plt.plot(bmi_september, 'r--',label = 'Bmi bulan September')
plt.title('Perbandingan Berat dan BMi bulan September', fontsize=10)
plt.xlabel('Peserta')
plt.ylabel('Nilai')
plt.legend()
plt.show()
```
